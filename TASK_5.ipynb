Load & Inspect Data

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('/content/traffic_accidents.csv')
df.head()
```

 Inspect

```python
df.info()
df.describe()
df.isnull().sum().sort_values(ascending=False).head(10)

```

CLEAN

```python
# Drop irrelevant/missing-heavy columns
df.drop(['X','Y',  'CRASH_RECORD_ID'], axis=1, inplace=True, errors='ignore')

# Drop rows missing key info
df.dropna(subset=['crash_date'], inplace=True)
```

cleaned the data, you can convert the crash_date column into datetime and extract time features:

```python
df['crash_date'] = pd.to_datetime(df['crash_date'])

# Extract hour and weekday for plotting
df['Hour'] = df['crash_date'].dt.hour
df['DayOfWeek'] = df['crash_date'].dt.day_name()
```

Accidents by Hour of Day (Rush hour analysis) Visualize Key Trends

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.countplot(x='Hour', data=df, palette='viridis')
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour (0 = Midnight, 23 = 11PM)')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

 Accidents by Day of Week

```python
plt.figure(figsize=(10,5))
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(x='DayOfWeek', data=df, order=day_order, palette='plasma')
plt.title('Accidents by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
